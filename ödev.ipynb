```python
from datetime import datetime

class banka_hesabi:
    def __init__(self,hesap_no,bakiye,hesap_turu):
        self.hesap_no = hesap_no
        self.bakiye = bakiye
        self.hesap_turu = hesap_turu
        self.hesap_tarihi=datetime.now()


        self.islem_gecmisi=[]
        self.islem_gecmisi.append(f"{self.datetime.now()}-Hesap oluşturuldu.Başlangıç bakiyesi {bakiye} TL")

    def __str__(self):
        return f"{self.hesap_no} nolu hesap – Bakiye: {self.bakiye} TL – Tür: {self.hesap_turu}"

    def __add__(self, other):
        return f"Bakiye={self.bakiye}+{other.bakiye}"

    def para_yatir(self, miktar):
        self.bakiye += miktar
        self.islem_gecmisi.append(f"{datetime.now()} - {miktar} TL yatırıldı. Yeni bakiye: {self.bakiye}")

    def para_cek(self, miktar):
        if miktar > self.bakiye:
            print("Yetersiz bakiye!")
            self.islem_gecmisi.append(f"{datetime.now()} - {miktar} TL çekilmek istendi ancak YETERSİZ BAKİYE.")
        else:
            self.bakiye -= miktar
            self.islem_gecmisi.append(f"{datetime.now()} - {miktar} TL çekildi. Yeni bakiye: {self.bakiye}")

    def hesap_raporu(self):
        print("Hesap Raporu")
        for gecmis in self.islem_gecmisi:
            print(gecmis)


```

```python
from collections import Counter
import math

class Ogrenci:
    def __init__(self, ad, soyad, ogrenci_no):
        self.ad = ad
        self.soyad = soyad
        self.ogrenci_no = ogrenci_no


        self.ders_notlari = {}

    def not_ekle(self, ders_adi, not_listesi):

        self.ders_notlari[ders_adi] = not_listesi
        sayac = Counter(not_listesi)

        print(f"{ders_adi} dersi için not ekleme tamamlandı.")
        print("Not istatistikleri (Counter):", sayac)

    def ortalama_hesapla(self, ders_adi):
        if ders_adi not in self.ders_notlari:
            print("Bu ders bulunamadı!")
            return None

        notlar = self.ders_notlari[ders_adi]
        ort = sum(notlar) / len(notlar)

        # math.floor - aşağı yuvarlama
        return math.floor(ort)

class LisansOgrencisi(Ogrenci):
    def ortalama_hesapla(self, ders_adi):
        if ders_adi not in self.ders_notlari:
            print("Ders bulunamadı!")
            return None

        notlar = self.ders_notlari[ders_adi]
        ort = sum(notlar) / len(notlar)


        return math.ceil(ort + 5)

class YuksekLisansOgrencisi(Ogrenci):
    def ortalama_hesapla(self, ders_adi):
        if ders_adi not in self.ders_notlari:
            print("Ders bulunamadı!")
            return None

        notlar = self.ders_notlari[ders_adi]
        ort = sum(notlar) / len(notlar)

        if ort >= 70:
            return f"Ortalama: {ort:.2f} → GEÇTİ"
        else:
            return f"Ortalama: {ort:.2f} → KALDI"
```

```python
import os
import random
import re
from datetime import datetime

class DosyaYoneticisi:
    def __init__(self, klasor):
        self.klasor = klasor

    def dosya_olustur(self, icerik=""):
        isim = str(random.randint(10000, 99999)) + ".txt"
        tam_yol = os.path.join(self.klasor, isim)

        with open(tam_yol, "w", encoding="utf-8") as f:
            f.write(icerik)

        print(f"{isim} oluşturuldu.")
        return tam_yol

    def dosya_oku_regex(self, dosya_yolu, pattern):
        with open(dosya_yolu, "r", encoding="utf-8") as f:
            metin = f.read()

        bulunanlar = re.findall(pattern, metin)
        print(f"Bulunan eşleşmeler: {bulunanlar}")
        return bulunanlar

    def klasor_tarama(self):
        dosyalar = os.listdir(self.klasor)
        print("Klasördeki dosyalar:", dosyalar)
        return dosyalar

    def __len__(self):
        return len(os.listdir(self.klasor))
```

```python
import random
from datetime import datetime
import math


class Urun:
    def __init__(self, isim, fiyat, stok):
        self.isim = isim
        self.fiyat = fiyat
        self._stok = stok
        self.barkod = random.randint(10000000, 99999999)

    @property
    def stok(self):
        return self._stok

    @stok.setter
    def stok(self, yeni_stok):
        if yeni_stok < 0:
            raise ValueError("Stok negatif olamaz!")
        self._stok = yeni_stok


class Musteri:
    def __init__(self, ad, email):
        self.ad = ad
        self.email = email
        self.sepet = []
        self.siparisler = []

    def sepete_ekle(self, urun):
        self.sepet.append(urun)

    def __contains__(self, urun):
        return urun in self.sepet


class Siparis:
    def __init__(self, urun_listesi):
        self.no = random.randint(100000, 999999)
        self.tarih = datetime.now()
        self.urunler = urun_listesi
        self.toplam_tutar = sum(u.fiyat for u in urun_listesi)

    def odeme_yap(self):
        return round(self.toplam_tutar * 1.18, 2)


def stok_guncelle(urun_listesi):
    for u in urun_listesi:
        u.stok -= 1
```

```python
import random
import time
from datetime import datetime
from collections import Counter
import math


class Quiz:
    def __init__(self):

        self.soru_bankasi = {
            1: {"soru": "Python hangi dil türüdür?", "cevap": "yorumlanan", "zorluk": 1},
            2: {"soru": "OOP açılımı nedir?", "cevap": "nesne tabanlı programlama", "zorluk": 2},
            3: {"soru": "Lambda fonksiyonu ne işe yarar?", "cevap": "anonim fonksiyon", "zorluk": 2},
            4: {"soru": "PEP8 nedir?", "cevap": "stil rehberi", "zorluk": 3},
            5: {"soru": "List comprehension ne sağlar?", "cevap": "kısa liste üretimi", "zorluk": 1},
            6: {"soru": "Python'da mutable veri tipinin örneği nedir?",
               "cevap": "liste","zorluk": 1},
            7: {"soru": "Varsayılan parametre değerleri Python’da ne zaman değerlendirilir?",
                "cevap": "fonksiyon tanımlandığı anda","zorluk": 3},
        }

        self.secilen_sorular = []
        self.kullanici_cevaplari = []
        self.istatistik = Counter()

    def soru_sec(self, adet):
        self.secilen_sorular = random.sample(list(self.soru_bankasi.values()), adet)

    def quiz_baslat(self):
        print("\n--- QUIZ BAŞLAgDI ---")
        baslangic = time.time()

        for i, soru in enumerate(self.secilen_sorular, start=1):
            print(f"\nSoru {i}: {soru['soru']}")
            cevap = input("Cevabınız: ").lower().strip()
            self.kullanici_cevaplari.append((soru, cevap))

        bitis = time.time()
        sure = round(bitis - baslangic, 2)
        print(f"\nToplam süre: {sure} saniye")
        return sure

    def cevap_kontrol(self):
        dogru = 0
        yanlis = 0

        for soru, cevap in self.kullanici_cevaplari:
            if cevap == soru["cevap"]:
                dogru += 1
            else:
                yanlis += 1

        self.istatistik = Counter({"Doğru": dogru, "Yanlış": yanlis})
        return self.istatistik

    def puan_hesapla(self):
        toplam_puan = 0
        for soru, cevap in self.kullanici_cevaplari:
            if cevap == soru["cevap"]:
                toplam_puan += soru["zorluk"] * 10  # zorluk katsayısı

        return toplam_puan

    def rapor_olustur(self, sure):
        print("\n--- QUIZ RAPORU ---")
        print("Tarih:", datetime.now())
        print("Doğru / Yanlış:", self.istatistik)
        print("Toplam Süre:", sure, "sn")
        print("Puan:", self.puan_hesapla())
```

```python
from datetime import datetime

class Personel:
    def __init__(self, ad, soyad, maas, departman, ise_baslama):
        self.ad = ad
        self.soyad = soyad
        self.maas = maas
        self.departman = departman
        self.ise_baslama = datetime.strptime(ise_baslama, "%Y-%m-%d")

    def zam_hesapla(self):
        gun = (datetime.now() - self.ise_baslama).days
        yil = gun // 365
        return round(self.maas * (1 + yil * 0.10), 2)

    def izin_hesapla(self):
        gun = (datetime.now() - self.ise_baslama).days
        return gun // 30

    def __eq__(self, other):
        return self.maas == other.maas

    def __gt__(self, other):
        return self.ise_baslama < other.ise_baslama


class Yonetici(Personel):
    def __init__(self, ad, soyad, maas, departman, ise_baslama):
        super().__init__(ad, soyad, maas, departman, ise_baslama)
        self.sorumlular = []

    def bonus(self):
        return len(self.sorumlular) * 500


class Gelistirici(Personel):
    def __init__(self, ad, soyad, maas, departman, ise_baslama):
        super().__init__(ad, soyad, maas, departman, ise_baslama)
        self.teknolojiler = []

    def prim(self):
        return len(self.teknolojiler) * 1000
```
